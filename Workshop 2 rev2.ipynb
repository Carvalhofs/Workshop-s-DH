```python
for i in df_enem[['Q001', 'Q002', 'Q003', 'Q004', 'Q005', 'Q006', 'Q007', 'Q008', 'Q009',
       'Q010', 'Q011', 'Q012', 'Q013', 'Q014', 'Q015', 'Q016', 'Q017', 'Q018',
       'Q019', 'Q020', 'Q021', 'Q022', 'Q023', 'Q024', 'Q025']]:
    print(df_enem[i].unique())
```

```python
df_enem.TP_SEXO = df_enem.TP_SEXO.replace({'M':0, 'F':1})
valores = {'':99,'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'H':8, 'I':9, 'J':10, 'K':11, 'L':12, 'M':13, 'N':14, 'O':15, 'P':16, 'Q':17, 'R':18, 'S':19, 'T':20}
for i in ['Q0'+'{:02}'.format(i) for i in range(1,26)]:
  df_enem[i] = df_enem[i].replace(valores)
```

```python
df_enem['media_notas'] = (df_enem['NU_NOTA_CH'] + df_enem['NU_NOTA_CN'] + df_enem['NU_NOTA_MT'] + df_enem['NU_NOTA_LC'] + df_enem['NU_NOTA_REDACAO']) / 5
```

```python
X = df_enem[['TP_FAIXA_ETARIA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO', 'TP_ANO_CONCLUIU',
       'TP_ESCOLA', 'TP_ENSINO', 'IN_TREINEIRO', 'TP_LINGUA', 'TP_STATUS_REDACAO',
       'Q001', 'Q002', 'Q003', 'Q004', 'Q005', 'Q006', 'Q007', 'Q008', 'Q009',
       'Q011', 'Q012', 'Q013', 'Q014', 'Q016', 'Q017', 'Q018',
       'Q019', 'Q020', 'Q022', 'Q023', 'Q024', 'Q025']]
```

```python
y = df_enem['media_notas']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.dtypes)
print(y_train.dtypes)
```

```python
import statsmodels.api as sm
ols_regressor = sm.OLS(y_train, X_train, missing = "drop")
model = ols_regressor.fit()
```

```python
pred_X_test = model.predict(X_test)
print(pred_X_test)
```
